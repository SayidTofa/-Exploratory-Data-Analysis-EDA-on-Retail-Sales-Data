```python
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
```

```python
df["Total Amount"].sum()
```

```python
df['Calculated Total'] = df['Quantity'] * df['Price per Unit']
# print("Total Amount matches Calculated Total:", (df['Total Amount'] == df['Calculated Total']).all())
```

```python
df.head()
```

```python
df_groupby = df.groupby("Product Category")["Gender"].value_counts().plot(kind="bar")
plt.ylabel("Total Products Count")
plt.title("Products Category by Gender")
# plt.figure("figsize=(8, 5))
plt.figure(figsize=(8, 5)) 
df[["Product Category","Gender"]].value_counts()
```

```python
df_groupby = df.groupby("Product Category")["Total Amount"].sum().plot(kind="bar", color=["pink","blue","purple"])
plt.ylabel("Total Amount")
plt.title("Product Category by Total Amount Sold")
plt.figure(figsize=(8, 5)) 
```

```python
df["Product Category"].value_counts()
```

```python
df["Product Category"].value_counts().plot(kind="bar",color=["blue","purple","pink"])
plt.ylabel("Total Sales")
plt.title("Total Sales by Product Category")
# df[["Product Category","Total Amount"]].value_counts()
```

```python
df["Gender"].value_counts().plot(kind="bar", color=["green", "brown"])
plt.title("Gender Count")
df["Gender"].value_counts()
```

```python
sales_by_gender = df.groupby("Gender")["Total Amount"].sum()
sales_by_gender.plot(kind="bar",color=["green","brown"])
plt.title("Total Sales by Gender")
plt.ylabel("Total Sales")
# df[["Total Amount", "Gender"]].value_counts()

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df["Month"] = df["Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum().plot.line()
plt.title("Total Sales by Month")
plt.ylabel("Total Sales")
```

```python
df["Year"] = df["Date"].dt.year
```

```python
Yearly_Sales = df.groupby("Year")["Total Amount"].sum().plot(kind="bar",color=["green","red"])
```
